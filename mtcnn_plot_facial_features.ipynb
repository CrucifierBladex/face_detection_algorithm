```
img_path=r'/content/tom.jpg'
```

```
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
image=cv2.imread(img_path)
detector=MTCNN()
out=detector.detect_faces(image)
x,y,w,h=out[0]['box']


def plot(img):

  cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
  plt.figure(figsize=(12,10))
  plt.imshow(image,cmap='gray')
  left_eye=list(out[0]['keypoints']['left_eye'])
  right_eye=list(out[0]['keypoints']['right_eye'])
  nose=list(out[0]['keypoints']['nose'])
  mouth_left=list(out[0]['keypoints']['mouth_left'])
  mouth_right=list(out[0]['keypoints']['mouth_right'])
  x1,y1=left_eye
  x2,y2=right_eye
  x3,y3=nose
  x4,y4=mouth_left
  x5,y5=mouth_right
  plt.imshow(img)
  plt.plot(x1,y1,'rx')
  plt.plot(x2,y2,'rx')
  plt.plot(x3,y3,'rx')
  plt.plot(x4,y4,'*')
  plt.plot(x5,y5,'*')
  plt.show()

plot(image)  

```
